```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
import seaborn as sns;
import warnings;
warnings.filterwarnings('ignore');
```

```python
df=pd.read_csv("train.txt",sep=';',header=None, names=['Text','Emotion'])
```

```python
df.head()
```

```python
emotion_array=df['Emotion'].unique()
emotion_array
```

```python
a={}
i=0
for emo in emotion_array:
  a[emo]=i
  i+=1
df['Emotion']=df['Emotion'].map(a)
```

```python
df.head()
```

```python
df['Text']=df['Text'].apply(lambda x : x.lower())
```

```python
import string
def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['Text']=df['Text'].apply(remove_punc)
```

```python
def remove_num(txt):
  s=""
  for i in txt:
    if not i.isdigit():
      s+=i
  return s
```

```python
df['Text']=df['Text'].apply(remove_num)
```

```python
def remove_emojis(txt):
  new=""
  for i in txt:
    if i.isascii():
      new+=i
  return new
```

```python
df['Text']=df['Text'].apply(remove_emojis)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words=set(stopwords.words('english'))
```

```python
def remove_stopwords(txt):
  s=txt.split()
  a=""
  for i in s:
    if i not in stop_words:
      a+=i+" "
  return a
```

```python
remove_stopwords(df.loc[1]['Text']);
```

```python
df['Text']=df['Text'].apply(remove_stopwords);
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df['Text'],df['Emotion'],test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer

bow_vectorization=CountVectorizer();
```

```python
X_train_bow=bow_vectorization.fit_transform(X_train)
X_test_bow=bow_vectorization.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
nb_model=MultinomialNB()
nb_model.fit(X_train_bow,y_train)
```

```python
y_pred=nb_model.predict(X_test_bow)
print("Accuracy Score of Bag of Words",accuracy_score(y_test,y_pred))
```

```python
tfid_vectorization=TfidfVectorizer();
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df['Text'],df['Emotion'],test_size=0.2,random_state=42)
```

```python
X_train_tfid=tfid_vectorization.fit_transform(X_train)
X_test_tfid=tfid_vectorization.transform(X_test)
```

```python
nb_model1=MultinomialNB()
nb_model1.fit(X_train_tfid,y_train)
```

```python
y_pred1=nb_model1.predict(X_test_tfid)
print("Accuracy Score of TF-IDF",accuracy_score(y_test,y_pred1));
```

```python
from sklearn.linear_model import LogisticRegression

lr_model=LogisticRegression(max_iter=100)
lr_model.fit(X_train_tfid,y_train)

y_pred2=lr_model.predict(X_test_tfid)
print("Accuracy Score of Logistic Regression",accuracy_score(y_test,y_pred2))
```
